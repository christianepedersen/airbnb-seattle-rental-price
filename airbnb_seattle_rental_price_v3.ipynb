### An exploration of airbnb listings from October 11 2020 in Seattle, WA

##### Analysis Questions:
- Can we accurately predict the price of an airbnb rental based on features of its airbnb post
- Which of these rental features are the most important in determining the price?

##### Analysis Plan:
- Optimize linear regression model
- Apply L1 (Lasso) regularization to reduce less important features
- Optimize XGboost regression model
- Use XGboost tree to examine which features of airbnbs are driving their price
- Are 'overpriced' airbnbs still being booked routinely?

### Regression model to predict price from features

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import preprocessing
```

```python
# pick out a subset of features for regression model

temp4_df = temp3_df[['price','zipcode','property_type','accommodates','bathrooms','bedrooms',
                     'beds','bed_type','guests_included','review_scores_rating','instant_bookable',
                     'availability_60']]
```

```python
# identify categorical variables for one-hot

# zipcode
print(temp4_df['zipcode'].unique())

# property_type
print(temp4_df['property_type'].unique())

# bed_type
print(temp4_df['bed_type'].unique())

# bed_type
print(temp4_df['instant_bookable'].unique())
```

```python
# one-hot encoding of categorical variables

temp4_df = pd.get_dummies(temp4_df, columns=['zipcode','property_type','bed_type','instant_bookable'])

del temp4_df['instant_bookable_f']
```

```python
print(temp4_df.shape)
temp4_df.head(5)
```

```python
# Drop rows that contain null values

temp4_df = temp4_df.dropna()
print(temp4_df.shape)
temp4_df.describe()
```

```python
# drop availability_60 and save for later
booked_df = pd.DataFrame()
booked_df = temp4_df['availability_60'].values

del temp4_df['availability_60']
```

```python
# min-max scale features

x = temp4_df.iloc[:,1:].values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
input_df = pd.DataFrame(x_scaled)
target_df = temp4_df['price']
```

```python
# Lasso method of feature selection selects a subset of features will attenuating coefficients of 
# less informative features to zero.

# LASSO
from sklearn import linear_model

# define features and target variable (behavioral positive bin)
x_select, y_select = input_df, target_df

# alpha value is the penalty on features (how selective do we want to be?)
alpha = 0.1 # Increasing alpha can shrink more variable coefficients to 0
clf = linear_model.Lasso(alpha=alpha,max_iter=3000)
clf.fit(x_select, y_select)

print(clf.coef_)
print(clf.intercept_)
```

```python
# pick out most informative features from LASSO feature selection

# check coeff list matches # features in x_smote
print(input_df.shape)
print(clf.coef_.shape)

# flag feature indices for features with non-zero coefficients
keep_idx = [i for i, e in enumerate(clf.coef_) if e != 0]
print(keep_idx)

# make new dataframe with subset of features for low-dimension model building
sub_X = input_df.iloc[:,keep_idx]
print(sub_X.shape)
```

```python
# Before training our regression models, we will split our data set
# and only use 80% of the timebins to train and the remaining 20% for validation at the end

from sklearn.model_selection import train_test_split

# Separate Features from Target
X = sub_X.values   # load subselected features into X DF
#X = input_df.values    # load all features
Y = target_df.values   # Load target into Y DF

# split into a training (80%) and testing set (20%)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)

# predict price for test airbnb posts
y_pred = lr.predict(X_test)

# model evaluation
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# printing values
print('Slope:' ,np.round(lr.coef_,0))
print('Intercept:', lr.intercept_)
print('Root mean squared error: ', rmse)
print('R2 score: ', r2)
```

```python
# plotting values

# data points
plt.subplots(figsize=(9, 9))
plt.scatter(y_test, y_pred, s=10)
plt.xlabel('Real price')
plt.ylabel('Predicted price')
plt.xlim([0, 1000])
plt.ylim([0, 1000])

# predicted values
plt.plot(range(1000),range(1000), color='r')
plt.show()
```

```python
# categorize airbnb posts based on whether they charge differently than we modeled

resids = y_test - y_pred

# predict price of all rentals using lin reg model
y_pred_all = lr.predict(X)

price_diff = Y/y_pred_all
price_diff[price_diff>5] = 5
price_diff[price_diff<0] = 0

# plot histogram of residual ratio
plt.hist(price_diff, bins=50)
plt.xlabel('Real/Predicted Price')
plt.ylabel('Number of airbnbs')
plt.show()
```

#### Distribution of airbnb rental prices
Most Airbnb rentals are priced in a predictable way that fits our linear regression model. However, there is a distinct outlier population that has a much higher price than can be predicted by our standard accomodation metrics (location, rental_type, number of people accommodated). Perhaps this is due to these venues offering special experiences that are not generally included in most stays. Also, are these expensive outliers truly 'overpriced' or are they regularly booked despite their cost?

```python
# Scatter plot and correlation between price difference and booking %

print(price_diff.shape)
print(booked_df.shape)

# booked_df originally number of days/60 that are available
booked_perc = (60-booked_df)/60

# data points
plt.subplots(figsize=(12, 9))
plt.scatter(booked_perc,price_diff, s=6)
plt.ylabel('Real/Predicted price ratio')
plt.xlabel('% Time Booked (60 days)')
#plt.xlim([0, 1000])
#plt.ylim([0, 1000])
```

#### 'Overpriced' airbnbs are often still consistently booked.

We have initially generated a predicted rental prices for airbnbs based on many of their categorical (one hot encoded) and numerical features using LINEAR regression and L1 LASSO regularization for feature selection.

We then calculated the ratio between the real price and our linearly predicted price.
- ratio >> 1 means real price of rental is higher than linear model --> "Overpriced"
- ratio == 1 means real price is equivalent to prediction
- ratio << 1 means real price is lower than linear model --> "Underpriced"

Using this approach, we might expect that 'overpriced' airbnbs would not be consistently booked by customers.
However, it appears that many 'overpriced' airbnbs are still very consistently booked.

Accordingly, I will now build a new regression model using XGboost regression to fit the data and predict price with lower "bias". I will also use the results of this model to investigate which of our airbnb features are primarily driving the price.

^If this approach does not work, perhaps we may need to apply NLP to the airbnb rental description to see if consistently booked "overpriced" airbnbs all have a common feature in their description (ex: "X minute walk to Space Needle!")


```python
import xgboost
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# x = featurespace (flat), target_df target

x = temp4_df.iloc[:,1:]
#min_max_scaler = preprocessing.MinMaxScaler()
#x = min_max_scaler.fit_transform(x)
print(x.shape)
# Note: x is featurespace after removing nans, one hot encoding but before LASSO

y = temp4_df.iloc[:,0]
print(y.shape)
target_df.all() == y.all()
```

```python
# Test, train split (80% train, 20% test), random state set for reproducibility

random_state = 1

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True, random_state=random_state)
```

```python
# Use GridSearch to find best model parameters

model = xgboost.XGBRegressor()
params={
    'booster': ['gbtree'], #['dart'] - best but shap not supported, #['gbtree', 'gblinear', 'dart'],
    'objective': ['reg:gamma'], #['reg:squarederror', 'reg:squaredlogerror', 'reg:logistic', 'reg:gamma'],
    'max_depth': [7], #[3, 4, 5, 6, 7, 10, 15], #--
    'learning_rate': [0.05, 0.1, 0.15], #--
    'n_estimators': [300, 350, 400],
#     'min_child_weight': [1, 10], #++
#     'colsample_bytree': [0.8],
#     'subsample': [0.75], 
#     'reg_alpha': [0],
#     'reg_lambda': [2], 
#     'gamma' : [0],
}
cv = KFold(5, shuffle=True, random_state=random_state)
rs = GridSearchCV(
    model,
    params,
    cv=cv,
    scoring="neg_mean_squared_error",
    n_jobs=2,   # use 2 CPU cores for parallel processing
    verbose=10)

rs.fit( 
    x, 
    y, 
    verbose=False)
print('best params:', rs.best_params_)
```

```python
# best_params = {'objective': 'reg:gamma', 'max_depth': 6, 'n_estimators': 200}
best_params = {
    'booster': 'gbtree', 
    'objective': 'reg:gamma',
    'max_depth': 7,
    'learning_rate': 0.05,
    'n_estimators': 300,
}
```

```python
model = xgboost.XGBRegressor(**best_params)
history = model.fit(
    x_train, y_train,
    eval_set=[(x_test, y_test)],
    eval_metric=['mae', 'rmse'], 
    early_stopping_rounds=int(best_params.get('n_estimators', 100) * 0.1))
```

```python
# Error of model on training data during iterations of training

pd.DataFrame(history.evals_result()['validation_0']).plot()
```

```python
y_true = y_test
y_pred = model.predict(x_test)

print('RMSE: ', np.sqrt(mean_squared_error(y_true, y_pred)))
print('MAE: ', mean_absolute_error(y_true, y_pred))
```

```python
pd.DataFrame({
    'pred': y_pred,
    'real': y_true,
    'diff': np.abs(y_pred - y_true),
}).iloc[:50].plot.bar(figsize=(15,5))
```

#### XGBoost regression fits data much better than linear regression

XGBoost regression had a RMSE on test data of ~79 
Linear regression had a RMSE on test data of ~124

We also see that the distribution (histogram) of real/predicted price ratio is a much tighter normal distribution for XGBoost than for linear regression.

Altogether, the XGBoost model fit relationship between airbnb feature and airbnb price much better. We also see that the most important features for predicting airbnb price were the review score of the host, the number of people accomodated, the number of bathrooms and the property type. We also see that being instantly bookable has a large impact on price, meaning that hosts that make that option available can charge more for an equivalent property. It is also clear that the zipcode in which the airbnb is located also has a major impact on the rentable price of the airbnb; certain locations can command higher prices.
